```python
import numpy as np
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
train = pd.read_csv('twitter_train.csv')
test = pd.read_csv('twitter_test.csv')
```

```python
train.head()
```

```python
train = train.drop(columns = ['tweet_id','airline','name','airline_sentiment_gold','name','negativereason_gold','retweet_count','tweet_coord','tweet_location','user_timezone','tweet_created'])
```

```python
# train = train.text
```

```python
train.head()
```

```python
from nltk.corpus import stopwords
import string
stops = stopwords.words('english')
stops += string.punctuation
```

```python
train.head()
```

```python
# documents = [([word for word in tokenizer.tokenize(text)],sent) for sent,text in train.iloc[:,:]]
documents = []
for index in range(len(train.iloc[:,0])) :
    text = train.iloc[index,1]
    sent = train.iloc[index,0]
    text = list(word_tokenize(text))
    new = []
    for word in text:
        if '@' in word:
            continue
        if len(word) < 3:
            continue
        if word in stops:
            continue
        else:
#             print(word)
            cleaned = lemmatizer.lemmatize(word.lower())
#             print(cleaned)
            new.append(cleaned)
    documents.append((new,sent))
```

```python
train_docs = documents
len(train_docs)
```

```python
test = test.drop(columns = ['tweet_id','airline','name','airline_sentiment_gold','name','negativereason_gold','retweet_count','tweet_coord','tweet_location','user_timezone','tweet_created'])
```

```python
test.head()
```

```python
all_words = []
for doc in train_docs:
    all_words += doc[0]

freq = nltk.FreqDist(all_words)
common = freq.most_common(1000)
features = [i[0] for i in common]
```

```python
features
```

```python
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
# training_data = [(get_word_set(text),sent) for text,sent in train_docs]
training_data = []
for text,sent in train_docs:
#     print(text,' ',sent)
    training_data.append((get_feature_dict(text),sent))
```

```python
test.head()
```

```python
# documents = [([word for word in tokenizer.tokenize(text)],sent) for sent,text in train.iloc[:,:]]
test_docs = []
for index in range(len(test.iloc[:,0])) :
    text = test.values[index][0]
#     sent = train.iloc[index,0]
    text = list(word_tokenize(text))
    new = []
    for word in text:
        if '@' in word:
            continue
        if len(word) < 3:
            continue
        if word in stops:
            continue
        else:
#             print(word)
            cleaned = lemmatizer.lemmatize(word.lower())
#             print(cleaned)
            new.append(cleaned)
    test_docs.append((new))
```

```python
test_docs
```

```python
# testing_data = [( (get_feature_dict(doc)) ) for doc in test_docs]
testing_data = []
for doc in test_docs:
    dic = get_feature_dict(doc)
    tup = ((dic))
    testing_data.append(tup)  
```

```python
testing_data[0]
```

```python
from nltk.classify.scikitlearn import SklearnClassifier
svc = SVC()
classifier_sklearn = SklearnClassifier(svc)
classifier_sklearn.train(training_data)
```
